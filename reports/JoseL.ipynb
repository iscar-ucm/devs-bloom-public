# Detección de outliers
Buscamos aquellos valores que se salen de los percentiles, y los reemplazamos por una interpolación, columna por columna.

```python
import pandas as pd
import numpy as np

df = pd.read_excel("../data/LatLon20210801.xlsx", sheet_name="Sheet1")
df.fillna(0, inplace=True)
print("df ANTES de la interpolación:")
print(df.head(30))

whisker_width = 1.5
for column in ['Lat','Lon']:
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_whisker = q1 - whisker_width*iqr
    upper_whisker = q3 + whisker_width*iqr
    df[column] = np.where(df[column] > upper_whisker, np.nan, df[column])
    df[column] = np.where(df[column] < lower_whisker, np.nan, df[column])
    df[column] = df[column].interpolate().ffill().bfill()
    print("df DESPUÉS de la interpolación de la columna " + column)
    print(df.head(30))
```
