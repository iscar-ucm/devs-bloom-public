```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt

vars=['DOX','NOX','ALG']
Sim=pd.read_csv("..\dataedge\Sensors2008_out.csv",parse_dates=True) 
```

```python



for var in vars:
    plt.figure(var)
    #plt.figure(figsize=(4, 4), dpi=300)
    plt.xlim(-122.23,-122.21)
    plt.ylim(47.5,47.521)
    plt.scatter(Sim.Lon,Sim.Lat,c=Sim[var], s=5, cmap=plt.cm.jet)
    cbar = plt.colorbar()
    #cbar.boundaries=[0,0.1]
    plt.xlabel('Longitude'), plt.xticks(rotation=30)
    plt.ylabel('Latitude')
    cbar.set_label(var)

    plt.title('Geographic coordinates')
    plt.gca().set_aspect('equal')
#plt.show()

```

```python
plt.rcParams["figure.figsize"] = (15,5)
myts=[dt.datetime.fromisoformat(s) for s in Sim['DateTime']]
for var in vars:
    plt.figure(var)
    #plt.figure(figsize=(4, 4), dpi=300)
    plt.plot(myts,Sim[var],'.-')
    plt.title(var)
    #plt.show()
```
