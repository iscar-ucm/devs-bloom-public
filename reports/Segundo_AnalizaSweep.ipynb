```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
varname = 'WQ_O'
varname = 'WQ_N'
varname='WQ_ALG'

Sim=pd.read_excel("..\dataedge\Sensors2008out2.xlsx",parse_dates=True) 

plt.figure(figsize=(4, 4), dpi=300)
plt.xlim(-122.35,-122.18)
plt.ylim(47.5,47.75)
plt.scatter(Sim.Lon,Sim.Lat,c=Sim[varname], s=5, cmap=plt.cm.jet)
cbar = plt.colorbar()
#cbar.boundaries=[0,0.1]
plt.xlabel('Longitude'), plt.xticks(rotation=30)
plt.ylabel('Latitude')
cbar.set_label(varname)

plt.title('Geographic coordinates')
plt.gca().set_aspect('equal')
plt.show()
```
